````python
from sqlalchemy import create_engine, MetaData, Table, text
import pandas as pd
````

````python
"""СОЗДАНИЕ БД И ТАБЛИЦЫ"""
import sqlite3
from sqlite3 import Error
def create_connection(db_file):
    """
    :return: Объект соединения или None
    """
    conn = None
    try:
        conn = sqlite3.connect(db_file)
        return conn
    except Error as e:
        print(e)

    return conn

def create_table(conn, sql_statement):
    try:
        crs = conn.cursor()
        crs.execute(sql_statement)
    except Error as e:
        print(e)

create_table_BikeShare = """
CREATE TABLE BikeShare (
    ride_id text PRIMARY KEY,
    rideable_type text,
    started_at text,
    ended_at text,
    start_station_name text,
    start_station_id integer,
    end_station_name text,
    end_station_id integer,
    start_lat real,
    start_lng real,
    end_lat real,
    end_lng real,
    member_casual text,
    duration integer);
"""        
        
conn = create_connection("C:/Users/206381/Downloads/Data/BikeshareServise.sqlite")
if conn is not None:
    crs = conn.cursor()
    crs.execute('DROP TABLE IF EXISTS BikeShare;')
    create_table(conn, create_table_BikeShare)
else:
    print("Ошибка! Не могу создать соединение с БД")
````

````python
#5.1. Покажите длительность с накопительным итогом по поездкам, начавшимся в интервале с 10:00 до 11:30 в порядке возрастания времени начала поездки.
stmt = """
select BikeShare.*, SUM(duration) OVER ( 
    ORDER BY started_at 
    ROWS BETWEEN UNBOUNDED PRECEDING AND CURRENT ROW),
    CAST(strftime('%H.%M', started_at) as float) as started_time
FROM BikeShare
WHERE started_time BETWEEN 10.00 AND 11.30
"""
 
df=pd.read_sql(stmt,conn)
df
````

````python
#5.2. 
stmt ="""SELECT start_station_name, END_station_name, DENSE_RANK () OVER (ORDER BY COUNT(ride_id) DESC)as rank,
COUNT(ride_id)
FROM BikeShare
GROUP BY start_station_name, end_station_name"""

df=pd.read_sql(stmt,conn)
df
````

````python
# 5.3. Покажите, сколько в проценках занимает время каждой поездки между 
# начальной и конечной станциями от суммарного времени, 
# затраченного на все поездки между станциями данной пары.

stmt = """select BikeShare.start_station_id, BikeShare.end_station_id, duration*100/sum(duration) OVER (
    PARTITION BY start_station_id, end_station_id) as perc
FROM BikeShare"""
df=pd.read_sql(stmt,conn)
df
````

````python
#5.4. Покажите 4 самых долгих поездки с каждой станции отправления упорядоченные
# по станции отправления и по убыванию длительности поездки для данной станции.
stmt = """
select * from 
(select BikeShare.start_station_id, duration, row_number() OVER (PARTITION BY start_station_id order by duration desc) as pr
FROM BikeShare
) as t1
where pr <=4
limit 20
"""
df=pd.read_sql(stmt,conn)
df
````

````python
#5.5. 
stmt = """
with t1 as (
select ride_id, end_station_name, end_station_id, duration
from bikeshare
order by duration desc
),
 
t2 as (
select t1.*,
NTILE (4) OVER (ORDER BY t1.duration) as RowNum
from t1
)
 
select t2.*
from t2
"""
df=pd.read_sql(stmt,conn)
df
````
